### 1. Create the following data frames based on the format provided: Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas

```python
# Import necessary libraries for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the downloaded Excel file
file_path = "board2.xlsx"  # Update with the actual path to the file
df = pd.read_excel(file_path)
df
```

#### a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# Filter rows where Track is 'Instrumentation', Hometown is 'Luzon', and Electronics score > 70
instru_df = df[(df['Track'] == 'Instrumentation') & 
                (df['Hometown'] == 'Luzon') & 
                (df['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']]

# Add constant columns 'Track' and 'Hometown' for clearness
instru_df['Track'] = 'Instrumentation'
instru_df['Hometown'] = 'Luzon'

instru_df
```

#### b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# b. Create DataFrame for Female Students in Mindanao with Average Grade >= 55
# Calculate the Average grade for each student
df['Average'] = (df['Math'] + df['GEAS'] + df['Electronics']) / 3

# Filter rows where Hometown is 'Mindanao', Gender is 'Female', and Average grade >= 55
mindy_df = df[(df['Hometown'] == 'Mindanao') & 
               (df['Gender'] == 'Female') & 
               (df['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]

# Add constant columns 'Gender' and 'Hometown' for clearness
mindy_df['Gender'] = 'Female'
mindy_df['Hometown'] = 'Mindanao'

mindy_df
```

### 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
# Create the figure and subplots
plt.figure(figsize=(18, 6))

# Boxplot for Track
plt.subplot(1, 3, 1)
sns.boxplot(x='Track', y='Average', data=df, palette='Set2')
plt.title('Average Score by Track', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Boxplot for Gender
plt.subplot(1, 3, 2)
sns.boxplot(x='Gender', y='Average', data=df, palette='Set2')
plt.title('Average Score by Gender', fontsize=16)
plt.xticks(rotation=0, fontsize=12)  # No rotation needed for gender
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Boxplot for Hometown
plt.subplot(1, 3, 3)
sns.boxplot(x='Hometown', y='Average', data=df, palette='Set2')
plt.title('Average Score by Hometown', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
